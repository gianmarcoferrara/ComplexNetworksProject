```python
import pandas as pd

#2020
#Upload del file CSV 
df_2020 = pd.read_csv('/Users/gianmarcoferrara/ComplexNetworksProject/filtered_2020.csv')

# Convert the 'date' column to datetime format to ensure correct time-based operations
df_2020['date'] = pd.to_datetime(df_2020['date'])

# Sort the data by 'fips' and 'date' to calculate the difference in cases for each fips
df_2020 = df_2020.sort_values(by=['fips', 'date'])

# Compute the difference in 'cases' for each 'fips' compared to the previous date
df_2020['new_cases'] = df_2020.groupby('fips')['cases'].diff()

#Fill not defined values with zeros
df_2020['new_cases'].fillna(0, inplace=True)

#Drop datas where we have not defined fips
df_2020_cleaned = df_2020.dropna(subset = ["fips"])

print(df_2020_cleaned)

#New CSV file
df_2020_cleaned.to_csv('/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2020.csv')


#2021
#Upload del file CSV 
df_2021 = pd.read_csv('/Users/gianmarcoferrara/ComplexNetworksProject/filtered_2021.csv')

# Convert the 'date' column to datetime format to ensure correct time-based operations
df_2021['date'] = pd.to_datetime(df_2021['date'])

# Sort the data by 'fips' and 'date' to calculate the difference in cases for each fips
df_2021 = df_2021.sort_values(by=['fips', 'date'])

# Compute the difference in 'cases' for each 'fips' compared to the previous date
df_2021['new_cases'] = df_2021.groupby('fips')['cases'].diff()

#Fill not defined values with zeros
df_2021['new_cases'].fillna(0, inplace=True)

#Drop datas where we have not defined fips
df_2021_cleaned = df_2021.dropna(subset = ["fips"])

#New CSV file
df_2021_cleaned.to_csv('/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2021.csv')


#2022
#Upload del file CSV 
df_2022 = pd.read_csv('/Users/gianmarcoferrara/ComplexNetworksProject/filtered_2022.csv')

# Convert the 'date' column to datetime format to ensure correct time-based operations
df_2022['date'] = pd.to_datetime(df_2022['date'])

# Sort the data by 'fips' and 'date' to calculate the difference in cases for each fips
df_2022 = df_2022.sort_values(by=['fips', 'date'])

# Compute the difference in 'cases' for each 'fips' compared to the previous date
df_2022['new_cases'] = df_2022.groupby('fips')['cases'].diff()

#Fill not defined values with zeros
df_2022['new_cases'].fillna(0, inplace=True)

#Drop datas where we have not defined fips
df_2022_cleaned = df_2022.dropna(subset = ["fips"])

#New CSV file
df_2022_cleaned.to_csv('/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2022.csv')


#2023
#Upload del file CSV 
df_2023 = pd.read_csv('/Users/gianmarcoferrara/ComplexNetworksProject/filtered_2023.csv')

# Convert the 'date' column to datetime format to ensure correct time-based operations
df_2023['date'] = pd.to_datetime(df_2023['date'])

# Sort the data by 'fips' and 'date' to calculate the difference in cases for each fips
df_2023 = df_2023.sort_values(by=['fips', 'date'])

# Compute the difference in 'cases' for each 'fips' compared to the previous date
df_2023['new_cases'] = df_2023.groupby('fips')['cases'].diff()

#Fill not defined values with zeros
df_2023['new_cases'].fillna(0, inplace=True)

#Drop datas where we have not defined fips
df_2023_cleaned = df_2023.dropna(subset = ["fips"])

#New CSV file
df_2023_cleaned.to_csv('/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2023.csv')
```
