```python
import os
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Path to the directory where the files are stored
directory = "/Users/Pietro/project/"  

# Step 1: Load the networks
network_list = []
for i in range(1, 33):  # 32 networks
    file_path = os.path.join(directory, f"adjacency_matrix_{i}.csv")
    try:
        # Load the adjacency matrix
        matrix = pd.read_csv(file_path, delimiter=',', header=None).values
        matrix = np.nan_to_num(matrix)  # Replace NaNs with 0
        G = nx.from_numpy_array(matrix)  # Create the graph
        network_list.append(G)  # Add to the list of networks
    except Exception as e:
        print(f"Error loading file {file_path}: {e}")

# Step 2: Lists for key parameters
densities = []
average_degrees = []
clustering_coefficients = []

# Step 3: Iterate and calculate parameters for each network
for i, G in enumerate(network_list, start=1):
    try:
        # Calculate parameters
        num_nodes = G.number_of_nodes()
        num_edges = G.number_of_edges()
        density = nx.density(G)
        avg_degree = sum(dict(G.degree()).values()) / num_nodes if num_nodes > 0 else 0
        avg_clustering = nx.average_clustering(G)
        
        # Save the results
        densities.append(density)
        average_degrees.append(avg_degree)
        clustering_coefficients.append(avg_clustering)
        
        # Print parameters for each network
        print(f"Network {i}:")
        print(f"  Number of nodes: {num_nodes}")
        print(f"  Number of edges: {num_edges}")
        print(f"  Density: {density}")
        print(f"  Average degree: {avg_degree}")
        print(f"  Average clustering coefficient: {avg_clustering}\n")
        
    except Exception as e:
        print(f"Error with network {i}: {e}")
        continue

# Step 4: Calculate and print the average values of each parameter across all networks
average_density = np.mean(densities)
average_degree = np.mean(average_degrees)
average_clustering = np.mean(clustering_coefficients)

print("\nAverage values across all networks:")
print(f"  Average Density: {average_density}")
print(f"  Average Degree: {average_degree}")
print(f"  Average Clustering Coefficient: {average_clustering}")

# Step 5: Visualize the parameters
plt.figure(figsize=(14, 8))

# Density
plt.subplot(2, 2, 1)
plt.plot(range(1, len(network_list) + 1), densities, marker='o', color='b')
plt.title("Network Density")
plt.xlabel("Network")
plt.ylabel("Density")
plt.grid(True)

# Average degree
plt.subplot(2, 2, 2)
plt.plot(range(1, len(network_list) + 1), average_degrees, marker='o', color='g')
plt.title("Average Degree")
plt.xlabel("Network")
plt.ylabel("Average Degree")
plt.grid(True)

# Clustering coefficient
plt.subplot(2, 2, 3)
plt.plot(range(1, len(network_list) + 1), clustering_coefficients, marker='o', color='r')
plt.title("Average Clustering Coefficient")
plt.xlabel("Network")
plt.ylabel("Clustering Coefficient")
plt.grid(True)

plt.tight_layout()
plt.show()
```
