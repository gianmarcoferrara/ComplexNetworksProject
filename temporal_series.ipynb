```python
import pandas as pd

file_paths = ['/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2020.csv',
              '/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2021.csv',
              '/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2022.csv',
              '/Users/gianmarcoferrara/ComplexNetworksProject/new_daily_2023.csv']

# Carica e unisce tutti i file CSV in un unico DataFrame
df_list = [pd.read_csv(file) for file in file_paths]
df_final = pd.concat(df_list)

# Assicurati che la colonna "date" sia di tipo datetime
df_final['date'] = pd.to_datetime(df_final['date'])

# Pivot del DataFrame: trasforma la colonna "contea" in colonne e riempi con "numero di nuovi casi"
df_pivot = df_final.pivot_table(index='date', columns='county', values='new_cases', aggfunc='sum')

# Riempire eventuali valori mancanti (NaN) con 0, se necessario
df_pivot = df_pivot.fillna(0)

subset1 = df_pivot.iloc[:20, :10]
print(subset1)

subset2 = df_pivot.iloc[-10:, -10:]
print(subset2)

subset3 = df_pivot.iloc[600:611, 1500:1511]
print(subset3)

# Salva il risultato su un nuovo file CSV
df_pivot.to_csv('serie_temporali_contee.csv')

print("Le serie temporali sono state salvate nel file 'serie_temporali_contee.csv'")
```
