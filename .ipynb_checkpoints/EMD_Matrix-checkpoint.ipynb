```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance

# File paths for your five distributions
file_paths = [
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_1.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_2.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_3.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_4.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_5.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_6.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_7.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_8.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_9.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_10.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_11.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_12.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_13.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_14.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_15.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_16.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_17.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_18.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_19.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_20.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_21.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_22.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_23.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_24.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_25.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_26.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_27.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_28.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_29.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_30.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_31.csv',
    '/Users/gianmarcoferrara/ComplexNetworksProject/divergencies/2_divergencies_32.csv'
]

# Load each distribution into a list
distributions = [pd.read_csv(file_path)['Divergence'] for file_path in file_paths]

# Initialize an empty matrix to store EMD values
num_distributions = len(distributions)
emd_matrix = np.zeros((num_distributions, num_distributions))

# Compute pairwise EMD
for i in range(num_distributions):
    for j in range(num_distributions):
        emd_matrix[i, j] = wasserstein_distance(distributions[i], distributions[j])

plt.figure(figsize=(10, 8))
sns.heatmap(
    emd_matrix,
    annot=True,
    fmt=".2f",
    cmap="viridis",  # Use a smoother colormap
    xticklabels=[f'Network {i+1}' for i in range(num_distributions)],
    yticklabels=[f'Network {i+1}' for i in range(num_distributions)],
    linewidths=0.5,  # Add a small gap for better distinction
    square=True
)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.title("Earth Mover's Distance (EMD) Between Networks")
plt.xlabel("Networks")
plt.ylabel("Networks")

# Save the figure before showing it
plt.savefig("2_emd_heatmap.png", dpi=300, bbox_inches='tight')

plt.show()
```
