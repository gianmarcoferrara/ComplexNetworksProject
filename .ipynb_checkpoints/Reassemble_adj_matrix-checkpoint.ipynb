```python
import pandas as pd
import numpy as np
from scipy.linalg import block_diag

#Diagonal blocks
sub_mat_1 = pd.read_csv("/Users/gianmarcoferrara/ComplexNetworksProject/31_sub_mat_1.csv", index_col=0)
sub_mat_2 = pd.read_csv("/Users/gianmarcoferrara/ComplexNetworksProject/31_sub_mat_2.csv", index_col=0)
sub_mat_3 = pd.read_csv("/Users/gianmarcoferrara/ComplexNetworksProject/31_sub_mat_3.csv", index_col=0)

#Convert each DataFrame to numpy arrays
sub_mat_1_np = sub_mat_1.values
sub_mat_2_np = sub_mat_2.values
sub_mat_3_np = sub_mat_3.values

# Convert each DataFrame to numpy arrays and set diagonal to zero
sub_mat_1_np = sub_mat_1.values
np.fill_diagonal(sub_mat_1_np, 0)

sub_mat_2_np = sub_mat_2.values
np.fill_diagonal(sub_mat_2_np, 0)

sub_mat_3_np = sub_mat_3.values
np.fill_diagonal(sub_mat_3_np, 0)

#Cross-matrices
cross_1_2 = pd.read_csv("/Users/gianmarcoferrara/ComplexNetworksProject/31_cross_matrix_1_2.csv", index_col=0)
cross_1_3 = pd.read_csv("/Users/gianmarcoferrara/ComplexNetworksProject/31_cross_matrix_1_3.csv", index_col=0)
cross_2_3 = pd.read_csv("/Users/gianmarcoferrara/ComplexNetworksProject/31_cross_matrix_2_3.csv", index_col=0)

#Convert each DataFrame to numpy arrays
cross_1_2_np = cross_1_2.values
cross_1_3_np = cross_1_3.values
cross_2_3_np = cross_2_3.values

cross_2_1_np = -cross_1_2_np.T
cross_3_1_np = -cross_1_3_np.T
cross_3_2_np = -cross_2_3_np.T

# Initialize a zero matrix of the final size: 3104 x 3104
full_matrix = np.zeros((3104, 3104))

# Place diagonal blocks
full_matrix[0:1000, 0:1000] = sub_mat_1_np
full_matrix[1000:2000, 1000:2000] = sub_mat_2_np
full_matrix[2000:3104, 2000:3104] = sub_mat_3_np

# Place off-diagonal blocks
full_matrix[0:1000, 1000:2000] = cross_1_2_np # top-right off-diagonal
full_matrix[0:1000, 2000:3104] = cross_1_3_np
full_matrix[1000:2000, 2000:3104] = cross_2_3_np

full_matrix[1000:2000, 0:1000] = cross_2_1_np  # top-right off-diagonal
full_matrix[2000:3104, 0:1000] = cross_3_1_np
full_matrix[2000:3104, 1000:2000] = cross_3_2_np

#Let's write the anti-symmetric matrix as in the paper
antisymm_full_mat_df = pd.DataFrame(full_matrix)
antisymm_full_mat_df.to_csv("31_antisymm_adjacency.csv")

# Set elements below the main diagonal to zero for networkX
full_matrix = np.triu(full_matrix)

# Calculate the total number of elements to keep (5% of ((3104*3104) - 3014) / 2)
total_elements = (3104 * 3104 - 3014) / 2
target_non_zero_count = int(total_elements * 0.05)
print("Target number of non-zero values to keep:", target_non_zero_count)

# Flatten the matrix and get the top values by absolute magnitude
non_zero_values = full_matrix[full_matrix != 0]
sorted_values_by_abs = np.sort(np.abs(non_zero_values))[::-1]  # Sort by absolute value in descending order

# Find the threshold value to keep only the top target_non_zero_count values by absolute value
if len(sorted_values_by_abs) > target_non_zero_count:
    threshold = sorted_values_by_abs[target_non_zero_count - 1]
else:
    threshold = sorted_values_by_abs[-1]  # If fewer non-zeros, take the smallest by absolute value

# Zero out values with absolute magnitude below the threshold
full_matrix[np.abs(full_matrix) < threshold] = 0

# Count the number of non-zero values after thresholding
non_zero_count_after = np.count_nonzero(full_matrix)
print("Number of non-zero values after thresholding:", non_zero_count_after)

# Create a new matrix with positive values set to zero
negative_only_matrix = np.copy(full_matrix)
negative_only_matrix[negative_only_matrix > 0] = 0

final_full_matrix = ((full_matrix - negative_only_matrix) - negative_only_matrix.T)

# Convert back to DataFrame for readability
final_full_matrix_df = pd.DataFrame(final_full_matrix)
final_full_matrix_df.to_csv("adjacency_matrix_31.csv")

percentage = non_zero_count_after/total_elements
print(percentage)

# Display the shape of the final matrix
print(final_full_matrix_df.shape)
```
